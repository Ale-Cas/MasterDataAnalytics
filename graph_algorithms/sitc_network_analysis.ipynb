# SITC Network Analysis

The data are a projection from a bipartite network of nations and the products they export.   
Edges weights represent a similarity score (called "proximity"). Data based on UN Comtrade worldwide trade patterns.   
SITC network based on the Standard International Trade Classification.  

In this notebook the focus will be on following points:
- a) Graphical representation
- b) Network classification
- c) Topological properties
- d) Connected components and their size distribution
- e) Node degrees and distribution 
- f) Most important vertex of network 
- g) Network transitivity level
- h) Community partition
- i) Assortativity degree
- l) Network randomness

```python
import igraph
import pandas as pd
import numpy as np
```

```python
edges_df = pd.read_csv('SITC/edges.csv', header=0, names=['source','target','weight'])
```

```python
edges_df.head()
```

## Graphical representation of the network

```python
graph = igraph.Graph.TupleList(edges_df[['source','target','weight']].itertuples(index=False), edge_attrs="weight")
```

```python
visual_style = {"bbox":(300,300), "margin":10}
igraph.plot(graph, vertex_color='blue', vertex_size=4, **visual_style)
```

## Network classification

```python
print(graph.summary())
```

From igrapgh classification Network can be defined as:
- U: undirected
- N: has vertex attribute called name
- W: weighted

## Basic topological properties of the network

```python
print('Number of vertices:', graph.vcount())
print('Number of edges:', graph.ecount())
print('Diameter:', graph.diameter())
print('Average path length:', graph.average_path_length())
```

```python
print('Average node degree:', np.mean(graph.degree()))
print('Maximum node degree:',max(graph.degree()))
print('Minumum node degree:',min(graph.degree()))
```
